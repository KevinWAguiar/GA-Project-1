#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
def std_dev(list_of_x):
    empty_list = []
    for i in list_of_x:
        new_number = ((float(i) - float(np.mean(list_of_x)))**2)
        empty_list.append(new_number)
    third_list = (sum(empty_list))/((len(list_of_x))-1)
    return third_list**(1/2)
````

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles):
    nrows = int(len(list_of_columns)/2)
    fig, axes = pyplt.subplots(figsize=(15,(2.75*len(list_of_titles))), nrows=nrows, ncols=2, sharey=True)
    axes = axes.ravel()
    for i, column in enumerate(list_of_columns):
        axes[i].set_title(list_of_titles[i], fontsize=12)
        axes[i].hist(dataframe[column]);
````

#### Plot and interpret boxplots

For each numeric variable in the dataframe create a boxplot using Seaborn. Boxplots demonstrate central tendency and spread in variables. In a certain sense, these are somewhat redundant with histograms, but you may be better able to identify clear outliers or differences in IQR, etc.

Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).

Each boxplot should:
- Only include variables of a similar scale
- Have clear labels for each variable
- Have appropriate titles and labels

````python
# Code
````

````python
def subplot_boxplot(dataframe, list_of_columns, list_of_titles):
    fig, ax = pyplt.subplots(figsize=(16, 5), nrows=1, ncols=3);
    ax[0].set_title((list_of_titles[0]));
    ax[1].set_title((list_of_titles[1]));
    
    ax[0].boxplot(dataframe[(list_of_columns)[0]]);
    ax[1].boxplot(dataframe[(list_of_columns)[1]]);
    
    sns.regplot(dataframe[(list_of_columns[0])], dataframe[(list_of_columns[1])]).set(xlabel=(list_of_titles[0]), ylabel=(list_of_titles[1]));

    pyplt.title("Correlation between: \n {} \n and \n {}".format((list_of_titles[0]), (list_of_titles[1])))
````
